```python
import math

class calc():
    def __init__(self,input1,input2):
        self.input1=input1
        self.input2=input2
    
    def addition(num1,num2):
        print(f"Addition = {num1+num2}")
        
    def subtraction(num1,num2):
        print(f"Subtraction = {num1-num2}")
        
    def multiplication(num1,num2):
        print(f"Multiplication = {num1*num2}")
        
    def division(num1,num2):
        print(f"Division = {num1/num2}")
         
    def area_of_rectangle(num1,num2):
        print(f"Area of Rectangle = {num1*num2}")
        
    def perimeter_of_rectangle(num1,num2):
        print(f"Perimeter of Rectangle = {2*(num1+num2)}")
        
    def squared(num1):
        print(f"Square = {num1**2}")
        
    def cubed(num1):
        print(f"Cube = {num1**3}")
        
    def perimeter_of_triangle(num1,num2,num3):
        print(f"Perimeter of Triangle = {num1+num2+num3}")
        
    def circumference_of_circle(num1):
        print(f"Circumference of Circle = {2*math.pi*num1}")
        
    def area_of_circle(num1):
        print(f"Area of Circle = {math.pi*(num1**2)}")
        
    def volume_of_sphere(num1):
        print(f"Volume of Sphere = {(4*math.pi*(num1**3))/3}")
        
    def surface_area_of_sphere(num1):
        print(f"Surface Area of Sphere = {4*math.pi*(num1**2)}")
        
    def average(num1,num2,num3):
        print(f"Average = {(num1+num2+num3)/3}")
        
    def area_of_square(num1):
        print(f"Area of Square = {4*num1}")
        
while True:
    print("Choose from the following menu: ")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Area of Rectangle")
    print("6. Perimeter of Rectangle")
    print("7. Square")
    print("8. Cube")
    print("9. Perimeter of Triangle")
    print("10. Circumference of Circle")
    print("11. Area of Circle")
    print("12. Volume of Sphere")
    print("13. Surface Area of Sphere")
    print("14. Average")
    print("15. Area of Square")
    print("16. Exit")
    
    choice=int(input("Enter your choice (1-16): "))
    
    if choice==1:
        num1=int(input("Enter the first number"))
        num2=int(input("Enter the second number"))
        calc.addition(num1,num2)
    elif choice==2:
        num1=int(input("Enter the first number"))
        num2=int(input("Enter the second number"))
        calc.subtraction(num1,num2)
    elif choice==3:
        num1=int(input("Enter the first number"))
        num2=int(input("Enter the second number"))
        calc.multiplication(num1,num2)
    elif choice==4:
        num1=int(input("Enter the first number"))
        num2=int(input("Enter the second number"))
        if num2==0:
            print("Infinity")
        else:
            calc.division(num1,num2)
    elif choice==5:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        calc.area_of_rectangle(num1,num2)
    elif choice==6:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        calc.perimeter_of_rectangle(num1,num2)    
    elif choice==7:
        num1=int(input("Enter a number: "))
        calc.squared(num1)   
    elif choice==8:
        num1=int(input("Enter a number: "))
        calc.cubed(num1)
    elif choice==9:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        num3=int(input("Enter the third number: "))
        calc.perimeter_of_triangle(num1,num2,num3)
    elif choice==10:
        num1=int(input("Enter the radius: "))
        calc.circumference_of_circle(num1)
    elif choice==11:
        num1=int(input("Enter the radius: "))
        calc.area_of_circle(num1)
    elif choice==12:
        num1=int(input("Enter the radius: "))
        calc.volume_of_sphere(num1)
    elif choice==13:
        num1=int(input("Enter the radius: "))
        calc.surface_area_of_sphere(num1)
    elif choice==14:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        num3=int(input("Enter the third number: "))
        calc.average(num1,num2,num3)
    elif choice==15:
        num1=int(input("Enter the first number: "))
        calc.area_of_square(num1)
    elif choice==16:
        break
    else:
        print("Your choice is not in the calculator")    
```
